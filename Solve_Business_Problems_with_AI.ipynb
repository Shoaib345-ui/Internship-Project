# Task 1. Classify emails

```python
import json

def classify_email(subject, message):
    prompt = f"""Classify the following email as either 'product inquiry' or 'order request'.
    Subject: {subject}
    Body: {message}
    Respond with only the category name."""

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    return response.choices[0].message.content.strip().lower().replace('"', '')

# Apply classification using 'message' column instead of 'body'
emails_df['category'] = emails_df.apply(lambda x: classify_email(x['subject'], x['message']), axis=1)
email_classification_df = emails_df[['email_id', 'category']].rename(columns={'email_id': 'email ID'})
display(email_classification_df.head())
```

# Task 2. Process order requests

```python
# Task 2: Order Processing and Response Generation

order_status_records = []
order_responses = []

def process_orders(row):
    if row['category'] != 'order request':
        return

    # Using 'message' column for the prompt
    prompt = f"""Extract product ID and quantity from this order request:
    {row['message']}
    Available Products: {products_df[['product_id', 'name']].to_dict()}
    Respond in JSON format: [{{'product_id': '...', 'quantity': int}}]"""

    # Placeholder for response generation logic
    response_text = f"Thank you for your order regarding the items mentioned in email {row['email_id']}. We have processed your request."
    order_responses.append({'email ID': row['email_id'], 'response': response_text})

# Re-run after classification is fixed
if 'category' in emails_df.columns:
    for index, row in emails_df[emails_df['category'] == 'order request'].iterrows():
        process_orders(row)

order_response_df = pd.DataFrame(order_responses)
display(order_response_df.head())
```

```python
# Task 2.2 & 3: Detailed Processing and RAG Inquiry Handling
import json

order_status_records = []
inquiry_responses = []

# Helper for Task 2.1: Stock check and Order Status
def process_stock_and_status(row):
    prompt = f"""Analyze this order: '{row['message']}'
    Catalog: {products_df[['product_id', 'name', 'stock']].to_dict('records')}
    Extract product_id and quantity. Check against stock.
    Return JSON list: [{{'product_id': str, 'quantity': int, 'status': 'created'|'out of stock'}}]"""
    res = client.chat.completions.create(model="gpt-4o", messages=[{"role": "user", "content": prompt}], response_format={"type": "json_object"})
    items = json.loads(res.choices[0].message.content).get('items', [])
    for item in items:
        item['email ID'] = row['email_id']
        order_status_records.append(item)

# Task 3: RAG Inquiry
def process_inquiry(row):
    # Simulate RAG retrieval from catalog
    context = products_df.to_string()[:2000]
    prompt = f"""Context: {context}\nInquiry: {row['message']}\nRespond professionally using the context."""
    res = client.chat.completions.create(model="gpt-4o", messages=[{"role": "user", "content": prompt}])
    inquiry_responses.append({'email ID': row['email_id'], 'response': res.choices[0].message.content})

# Run Processing
for _, row in emails_df.iterrows():
    if row['category'] == 'order request':
        process_stock_and_status(row)
    else:
        process_inquiry(row)

order_status_df = pd.DataFrame(order_status_records)
inquiry_response_df = pd.DataFrame(inquiry_responses)
display(order_status_df.head())
display(inquiry_response_df.head())
```

# Task 3. Handle product inquiry

```python
# Task 3: Handle Product Inquiry with RAG (Scaling to 100k+ products)
# For the POC, we simulate the retrieval by using LLM to find the most relevant product context

inquiry_responses = []

def handle_inquiry(row):
    if row['category'] != 'product inquiry':
        return

    # RAG: In a production system, we would query a Vector Store (e.g., Pinecone/FAISS) here.
    # For this task, we find relevant context from the products_df.
    relevant_products = products_df.to_string()[:3000] # Simplified RAG retrieval

    prompt = f"""Context: {relevant_products}\n\nQuestion: {row['message']}\n\nAnswer the customer inquiry professionally based on the context above."""

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )
    inquiry_responses.append({'email ID': row['email_id'], 'response': response.choices[0].message.content})

for index, row in emails_df[emails_df['category'] == 'product inquiry'].iterrows():
    handle_inquiry(row)

inquiry_response_df = pd.DataFrame(inquiry_responses)
display(inquiry_response_df.head())
```
